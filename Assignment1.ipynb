# 5.Perform meaningful visualisation using the dataset.

```python
df=data.groupby('Classification')['SL'].sum().to_frame().reset_index()
plt.bar(df['Classification'],df['SL'],color=['cornflowerblue','lightseagreen','steelblue'])
plt.title('Barchart')
plt.xlabel('Classification')
plt.ylabel('S L')
plt.show()
```

```python
df=data.groupby('Classification')['PL'].sum().to_frame().reset_index()
plt.bar(df['Classification'],df['PL'],color=['cornflowerblue','lightseagreen','steelblue'])
plt.title('Barchart')
plt.xlabel('Classification')
plt.ylabel('P L')
plt.show()
```

```python
col=data[['SL','SW','PL','PW']]
col
```

```python
plt.plot(col)
plt.title("line_chart")
plt.legend(col)
```

```python
fig,ax=plt.subplots()
ax.set_title('Boxplot')
ax.boxplot(col)
plt.show()
```
